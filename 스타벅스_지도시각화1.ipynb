```python
#json 형태 처리
# 지도시각화 라이브러리 불러오기
#conda install -c conda-forge folium
#conda install folium
#pip install folium
import folium
import json
```

```python
#최초에 보여줄 지도 위치[위도,경도]
starbucks_map = folium.Map(location = [37.573030, 126.979189],
#지도 스타일
#도시형 건물 등 일반적으로 사용
#tiles = 'openstreetmap',
#산림위주로 보여줌
#tiles = 'Stamen Terrain',
#하천이나 도로 위주로 보여줌
tiles = 'cartodbpositron',
#최초에 보여줄 짇 zoom 사이즈
zoom_start = 50
)
starbucks_map
```

```python
#스타벅스의 매장 위치를 표시하기
for idx in seoul_starbucks.index :
  lat = seoul_starbucks.loc[idx, '위도']
  ing = seoul_starbucks.loc[idx, '경도']
  #지도에 표시할 마킹 모양 표시하기
  folium.CircleMarker(location = [lat, ing], fill_color = 'green', fill_opacity=1, color = 'yellow', weight = 1, radius = 3).add_to(starbucks_map)

starbucks_map
```

```python
#매장 타입별로 마커 그리기
starbucks_map2 = folium.Map(
    location = [37.573030, 126.979189], tiles = 'openstreetmap',zoom_start = 11
)
for idx in seoul_starbucks.index : 
  lat = seoul_starbucks.loc[idx, '위도']
  lng = seoul_starbucks.loc[idx, '경도']
  store_type = seoul_starbucks.loc[idx, '매장타입']

  #매장 타입별 색상 선택을 위한 조건문 작성
  fillcolor = ''
  if store_type == 'general':
    fillColor= 'gray'
    size = 5
  elif store_type == 'reserve':
    fillColor= 'blue'
    size = 5
  elif store_type == 'generalDT':
    fillColor= 'red'
    size = 5

  folium.CircleMarker(
      location = [lat,lng], color = fillColor, fill=True, fill_color = fillColor, fill_opaciy = 1, weight = 1, radius = size
  ).add_to(starbucks_map2)
starbucks_map2
```
